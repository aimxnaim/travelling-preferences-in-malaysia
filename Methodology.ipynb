```python
# Read data from external file
import pandas as pd
dfCSV = pd.read_csv("travel_data_ori.csv")

dfCSV.info()
```

```python
dfCSV.head()
```

```python
#Report missing data
dfCSV.isna().sum()
```

```python
#Data transformation
from sklearn.preprocessing import LabelEncoder
var_mod = ['AGE', 'GENDER','MARITAL STATUS', 'Do you prefer solo or group travel?', 'What is your budget for travelling?', 'Usually where do you plan to stay?', 'Do you prefer using tour guide?', 'Do you prefer travelling local or overseas?']
le = LabelEncoder()
for i in var_mod:
    dfCSV[i] = le.fit_transform(dfCSV[i])


dfCSV.info()
```

```python
dfCSV.head()
```

```python
#changing income datatype to integer
import pandas as pd
dfCSV['AGE'] = pd.to_numeric(dfCSV['AGE'])

dfCSV.info()
```

```python
# Define the custom function to categorize the data
def categorize_jobs(job):
    if job in ["student","Student","students","Students","STUDENT","STUDENTS"]:
        return 0
    else:
        return 1

# Apply the function to the occupation column
dfCSV['OCCUPATION:'] =  dfCSV['OCCUPATION:'].apply(categorize_jobs)


# Check the result
dfCSV.info()
```

```python
dfCSV.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split


# split the data into training and testing sets
train_data, test_data = train_test_split(dfCSV, test_size=0.0)

# save the training set as a new CSV file
train_data.to_csv(f"travel_dataset_transform.csv", index=False)

# save the testing set as a new CSV file
test_data.to_csv(f"test-IT.csv", index=False)
```
